<br>12.Month: Month of the observation extracted from Date..

### TASK 1.3: Use the functionality of **Seaborn Library** to create a visualization to compare the sales trend per vehicle type for a recession period with a non-recession period.


<details><summary>Click here for a hint</summary>

<p>
     To visualize the average number of vehicles sold during recession and non-recession periods, you can use a bar chart
        <br> You will need to group recession average Automobile_Sales and then plot it<br>
    Make use of sns.barplot(x=x,y=y, data = df)</p>

````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your DataFrame is named df and contains 'Year', 'Automobile_Sales', and 'Vehicle_Type' columns

# Filter the data for recession and non-recession periods
recession_years = [1980, 1981, 1982, 1991, 2000, 2001, 2007, 2008, 2009, 2020]
recession_data = df[df['Year'].isin(recession_years)]
non_recession_data = df[~df['Year'].isin(recession_years)]

# Create a figure and axis for the plot
plt.figure(figsize=(12, 8))

# Plot the sales trend per vehicle type for recession period
sns.lineplot(data=recession_data, x='Year', y='Automobile_Sales', hue='Vehicle_Type', palette='tab10', markers=True, dashes=False)

# Set labels and title
plt.title('Sales Trend per Vehicle Type (Recession Period)')
plt.xlabel('Year')
plt.ylabel('Automobile Sales')
plt.xticks(rotation=90)

# Add a legend
plt.legend(title='Vehicle Type', loc='upper left')

# Create a second subplot for non-recession period
plt.figure(figsize=(12, 8))

# Plot the sales trend per vehicle type for non-recession period
sns.lineplot(data=non_recession_data, x='Year', y='Automobile_Sales', hue='Vehicle_Type', palette='tab10', markers=True, dashes=False)

# Set labels and title for the non-recession plot
plt.title('Sales Trend per Vehicle Type (Non-Recession Period)')
plt.xlabel('Year')
plt.ylabel('Automobile Sales')
plt.xticks(rotation=90)

# Add a legend for the non-recession plot
plt.legend(title='Vehicle Type', loc='upper left')

# Show both plots
plt.tight_layout()
plt.show()
````

### Now you want to compare the sales of different vehicle types during a recession and a non-recession period
<br>We recommend that you use the functionality of **Seaborn Library** to create this visualization


<details><summary>Click here for a hint</summary>

<p>
     To visualize sales of different vehicles during recession and non-recession periods, you can use a bar chart

````python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df and contains columns 'Year', 'Automobile_Sales', 'Vehicle_Type', and 'Recession'
# Filter the data for recession and non-recession periods
recession_years = [1980, 1981, 1982, 1991, 2000, 2001, 2007, 2008, 2009, 2020]
recession_data = df[df['Year'].isin(recession_years)]
non_recession_data = df[~df['Year'].isin(recession_years)]

# Create subplots with Seaborn
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharey=True)  # 1 row, 2 columns of subplots, sharing the y-axis

# Create bar plots for recession and non-recession periods
sns.barplot(x='Year', y='Automobile_Sales', hue='Vehicle_Type', data=recession_data, ax=axes[0])
sns.barplot(x='Year', y='Automobile_Sales', hue='Vehicle_Type', data=non_recession_data, ax=axes[1])

# Set titles and labels for subplots
axes[0].set_title('Automobile Sales by Vehicle Type (Recession Periods)')
axes[1].set_title('Automobile Sales by Vehicle Type (Non-Recession Periods)')
axes[0].set_xlabel('Year')
axes[1].set_xlabel('Year')
axes[0].set_ylabel('Automobile Sales')
axes[1].set_ylabel('')  # Remove y-axis label for the second subplot

# Rotate the year labels for better readability
axes[0].tick_params(axis='x', rotation=90)
axes[1].tick_params(axis='x', rotation=90)

# Display the plot
axes[0].grid(True)
axes[1].grid(True)
axes[0].legend(title='Vehicle Type', loc='upper right', bbox_to_anchor=(1.15, 1))
axes[1].legend(title='Vehicle Type', loc='upper right', bbox_to_anchor=(1.15, 1))
plt.tight_layout()  # Ensure proper spacing between subplots
plt.show()
````

### TASK 1.4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, you want to find more insights from the data to understand the reason. <br>Plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods? 
<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


````python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df and contains columns 'Year', 'GDP', and 'Recession'
# Filter the data for recession and non-recession periods
recession_years = [1980, 1981, 1982, 1991, 2000, 2001, 2007, 2008, 2009, 2020]
recession_data = df[df['Year'].isin(recession_years)]
non_recession_data = df[~df['Year'].isin(recession_years)]

# Create subplots for GDP during recession and non-recession periods
fig, axes = plt.subplots(2, 1, figsize=(12, 8), sharex=True, sharey=True)  # 2 rows, 1 column of subplots, sharing the x and y-axes

# Line plot for GDP during recession periods
axes[0].plot(recession_data['Year'], recession_data['GDP'], label='GDP during Recession', marker='o', color='red')
axes[0].set_title('GDP during Recession Periods')
axes[0].set_ylabel('GDP')
axes[0].grid(True)
axes[0].legend()

# Line plot for GDP during non-recession periods
axes[1].plot(non_recession_data['Year'], non_recession_data['GDP'], label='GDP during Non-Recession', marker='o', color='blue')
axes[1].set_title('GDP during Non-Recession Periods')
axes[1].set_xlabel('Year')
axes[1].set_ylabel('GDP')
axes[1].grid(True)
axes[1].legend()

# Display the subplots
plt.tight_layout()
plt.show()
````

````python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df and contains columns 'Month', 'Automobile_Sales', and 'Seasonality_Weight'
# Filter the data for non-recession years
recession_years = [1980, 1981, 1982, 1991, 2000, 2001, 2007, 2008, 2009, 2020]
non_recession_data = df[~df['Year'].isin(recession_years)]

# Create a bubble plot for displaying seasonality impact on automobile sales
plt.figure(figsize=(12, 8))

# Scatter plot with bubbles (size represents Seasonality Weight)
plt.scatter(
    non_recession_data['Month'],  # X-axis: Month
    non_recession_data['Automobile_Sales'],  # Y-axis: Automobile Sales
    s=non_recession_data['Seasonality_Weight'] * 100,  # Bubble size proportional to Seasonality Weight
    alpha=0.6,  # Transparency
    cmap='coolwarm',  # Color map
    c=non_recession_data['Automobile_Sales'],  # Color based on Automobile Sales
    label='Sales vs. Month'
)

# Set labels, title, and legend
plt.title('Seasonality Impact on Automobile Sales (Non-Recession Years)')
plt.xlabel('Month')
plt.ylabel('Automobile Sales')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.colorbar(label='Automobile Sales')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
````

### TASK 1.8: Develop a pie chart to display the total Advertisement expenditure for each vehicle type during recession period.<br>
Can we observe the share of each vehicle type in total sales during recessions? 
#### Create another pie plot to display the total advertisement expenditure for each vehicle type


<details><summary>Click here for a hint</summary>

<p>
     You will be required to group vehicle type for sum of advertisement expenditure.
    <br> the plot a pie with the data, May include relevant labels
    <br>title this plot as  - Share of Each Vehicle Type in Total Sales during Recessions</p>
</details>


````python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your DataFrame is named df and contains columns 'Recession', 'Automobile_Sales', 'Advertising_Expenditure', and 'Vehicle_Type'

# Filter the data for recession periods
recession_data = df[df['Recession'] == 1]

# Group the data by Vehicle_Type and calculate the total advertising expenditure for each type
ad_expenditure_by_type = recession_data.groupby('Vehicle_Type')['Advertising_Expenditure'].sum()

# Group the data by Vehicle_Type and calculate the total sales volume for each type
sales_volume_by_type = recession_data.groupby('Vehicle_Type')['Automobile_Sales'].sum()

# Create a pie chart to display the total advertising expenditure for each vehicle type during a recession
plt.figure(figsize=(12, 6))

# Pie chart for advertising expenditure
plt.subplot(121)
plt.pie(ad_expenditure_by_type, labels=ad_expenditure_by_type.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Advertising Expenditure by Vehicle Type (Recession)')

# Pie chart for sales volume during recession
plt.subplot(122)
plt.pie(sales_volume_by_type, labels=sales_volume_by_type.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Sales Volume by Vehicle Type (Recession)')

# Adjust the layout
plt.tight_layout()

# Show the pie charts
plt.show()
````
